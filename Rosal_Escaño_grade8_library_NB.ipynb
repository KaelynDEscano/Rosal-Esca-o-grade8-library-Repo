```python
import firebase_admin
from firebase_admin import credentials, db

# Check if the default Firebase app is already initialized
if not firebase_admin._apps: # _apps is a dictionary containing initialized apps
    #Load the private key
    cred = credentials.Certificate("/content/firebase_key.json")
    #Initialize the app with your database URL
    firebase_admin.initialize_app (cred, { "databaseURL": "https://rosal-escano-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"})
    print("Firebase connected successfully!")
else:
    print("Firebase app already initialized.")
```

```python
import json
with open("/content/books.json", "r") as file:
 data = json.load(file)
print("JSON file loaded")
ref = db.reference("books")
for books in data["books"]:
 ref.child(str(books["id"])).set(books)
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
total = 0
for book in data["books"]:
  print(f"Title: {book['title']}")
  print(f"Available: {book['details']['available']}")
  total += book['details']['copies']
print("Total number of books: " , total)
```
